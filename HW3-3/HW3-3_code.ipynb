# Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Load Images

```python
img1 = cv2.imread('checkerboard1024-shaded.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(img1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img2 = cv2.imread('N1.bmp', cv2.IMREAD_GRAYSCALE)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

# Useful Function

```python
def show_plots(image, shade, unmix):
    plt.figure(figsize=(12, 6))

    plt.subplot(131)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(132)
    plt.imshow(shade, cmap='gray')
    plt.title('Shading noise pattern')
    plt.axis('off')

    plt.subplot(133)
    plt.imshow(unmix, cmap='gray')
    plt.title('Final correct Image')
    plt.axis('off')

    plt.show()
```

### 1. checkerboard1024-shaded.tif

```python
img1_shading_patterns = cv2.GaussianBlur(img1, (257, 257), 64)
img1_divided = img1 / img1_shading_patterns
show_plots(img1, img1_shading_patterns, img1_divided)
```

### 2. N1.bmp

```python
img2_shading_patterns = cv2.GaussianBlur(img2, (99, 99), 64)
img2_divided = img2 / img2_shading_patterns
show_plots(img2, img2_shading_patterns, img2_divided)
```
